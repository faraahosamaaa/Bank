```python
import pandas as pd
```

```python
df=pd.read_csv("Bank_Customer.csv")
```

```python
df.head(30)
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=1,how='any')
```

```python
df.drop(df[(df['job'] == 'unknown') & (df['education'] == 'unknown')].index, inplace=True)
```

```python
df.info()
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['balance'] = pd.to_numeric(df['balance'], errors='coerce')
```

```python
Q1 = df['balance'].quantile(0.25)
Q3 = df['balance'].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df = df[(df['balance'] >= lower_bound) & (df['balance'] <= upper_bound)]
```

```python
average_age = df['age'].mean()
print(average_age)
```

```python
max_balance_per_job = df.groupby('job')['balance'].max().reset_index()
```

```python
max_balance_per_job
```

```python
duplicates = df.duplicated()
print("Number of duplicate rows: ", duplicates.sum())
```
